```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
```

```python
with open('mysmartprice.html','r',encoding = 'utf-8') as file:
    soup = BeautifulSoup(file, 'html.parser')
```

```python
containers = []
containers.extend(soup.find_all('div',{"class":"spec_card white_bg w-100 br-all br8 d-flex j-c pos-rel"})) #extract first phone
containers.extend(soup.find_all('div',{"class":"spec_card white_bg w-100 br-all br8 d-flex j-c pos-rel m-t-top"})) #extract all other phones
```

```python
item_no =1
name=[]
data = pd.DataFrame(columns=['mobile_name' ,'release_date','price','avg_rating','total_ratings','cpu','rear_camera','front_camera','display','ram_and_storage','battery_and_charging_speed','operating_system','5G|NFC|Fingerprint','expert_view' ])
for item in containers:
    
    print(item_no)
    item_no+=1

    #mobile name
    print(item.find('h2').text) 
    
    #release date
    try:
        print(item.find('div',{'class':'r_date gray_txt'}).text)
    except:
        print(np.nan)
        
    #price
    try:
        print(item.find('span',{'class':'price txt-w-b'}).text)
    except:
        print(np.nan)

    #average rating
    try:
        print(item.find("p", class_="r_value").text.strip().split('/')[0].split('\n')[1].strip())
    except:
        print(np.nan)
    
    #total ratings
    try:
        print(item.find("div", class_="rat_v ratng_bdr gray_txt").text.split('\n')[-2].strip())
    except:
        print(np.nan)

    #extracting specs
    #1. cpu
    try:
        print(item.find("li", class_=re.compile(".*cpu.*")).text.strip())
    except:
        print("processor not found")
    
    #2. rear camera
    try:
        print(item.find("li", class_=re.compile(".*cmra.*")).text.strip())
    except:
        print("rear camera not found")

    #3. Front camera
    try:
        print(item.find("li", class_=re.compile(".*front-cmra.*")).text.strip())
    except:
        print("front camera not found")
    
    #4. Display
    try:
        print(item.find("li", class_=re.compile(".*aspct.*")).text.strip())
    except:
        print("display not found")
    
    #5. Ram and Storage
    try:
        print(item.find("li", class_=re.compile(".*ram.*")).text.strip())
    except:
        print("ram not found")
    
    #6.battery and charging speed
    try:
        print(item.find("li", class_=re.compile(".*bttry.*")).text.strip())
    except:
        print("battery not found")

    #7.OS
    try:
        print(item.find("li", class_=re.compile(".*android-os.*")).text.strip())
    except:
        print("os not found")
    
    #8. 5G|NFC|Fingerprint
    try:
        print(item.find("li", class_=re.compile(".*5G.*")).text.strip())
    except:
        print("5G|NFC|Fingerprint not found")
    
    #9.expert view
    try:
        print(item.find("div", class_='vrfy_msp pos-rel trunk o_f_h').text.strip().split('\n')[0].strip())
    except:
        print('expert review not found')
    
    print()


    

    
```

```python
df = pd.DataFrame(columns=['mobile_name' ,'release_date','price','avg_rating','total_ratings','cpu','rear_camera','front_camera','display','ram_and_storage','battery_and_charging_speed','operating_system','5G|NFC|Fingerprint','expert_view' ])
data = {'mobile_name':[] ,'release_date':[],'price':[],'avg_rating':[],'total_ratings':[],'cpu':[],'rear_camera':[],'front_camera':[],'display':[],'ram_and_storage':[],'battery_and_charging_speed':[],'operating_system':[],'5G|NFC|Fingerprint':[],'expert_view':[]}
for item in containers:
    #mobile name
    data['mobile_name'].append(item.find('h2').text)
    
    #release date
    try:
        data['release_date'].append(item.find('div',{'class':'r_date gray_txt'}).text)
    except:
        data['release_date'].append(np.nan)
        
    #price
    try:
        data['price'].append(item.find('span',{'class':'price txt-w-b'}).text)
    except:
        data['price'].append(np.nan)

    #average rating
    try:
        data['avg_rating'].append(item.find("p", class_="r_value").text.strip().split('/')[0].split('\n')[1].strip())
    except:
        data['avg_rating'].append(np.nan)
    
    #total ratings
    try:
        data['total_ratings'].append(item.find("div", class_="rat_v ratng_bdr gray_txt").text.split('\n')[-2].strip())
    except:
        data['total_ratings'].append(np.nan)

    #extracting specs
    #1. cpu
    try:
        data['cpu'].append(item.find("li", class_=re.compile(".*cpu.*")).text.strip())
    except:
        data['cpu'].append(np.nan)
    
    #2. rear camera
    try:
        data['rear_camera'].append(item.find("li", class_=re.compile(".*cmra.*")).text.strip())
    except:
        data['rear_camera'].append(np.nan)

    #3. Front camera
    try:
        data['front_camera'].append(item.find("li", class_=re.compile(".*front-cmra.*")).text.strip())
    except:
        data['front_camera'].append(np.nan)
    
    #4. Display
    try:
        data['display'].append(item.find("li", class_=re.compile(".*aspct.*")).text.strip())
    except:
        data['display'].append(np.nan)
    
    #5. Ram and Storage
    try:
        data['ram_and_storage'].append(item.find("li", class_=re.compile(".*ram.*")).text.strip())
    except:
        data['ram_and_storage'].append(np.nan)
    
    #6.battery and charging speed
    try:
        data['battery_and_charging_speed'].append(item.find("li", class_=re.compile(".*bttry.*")).text.strip())
    except:
        data['battery_and_charging_speed'].append(np.nan)

    #7.OS
    try:
        data['operating_system'].append(item.find("li", class_=re.compile(".*android-os.*")).text.strip())
    except:
        data['operating_system'].append(np.nan)
    
    #8. 5G|NFC|Fingerprint
    try:
        data['5G|NFC|Fingerprint'].append(item.find("li", class_=re.compile(".*5G.*")).text.strip())
    except:
        data['5G|NFC|Fingerprint'].append(np.nan)
    
    #9.expert view
    try:
        data['expert_view'].append(item.find("div", class_='vrfy_msp pos-rel trunk o_f_h').text.strip().split('\n')[0].strip())
    except:
        data['expert_view'].append(np.nan)
```
